# Оценка ПК для современных игр и комфортного игрового опыта

## 1. Параметры для оценки и их важность

При оценке ПК для современных игр учитываются следующие ключевые параметры:

- **Процессор (CPU)**  
  Отвечает за обработку игровой физики, ИИ и многозадачность.  
  Минимум: 6 ядер (Intel i5-12400F / AMD Ryzen 5 5600X и новее).

- **Видеокарта (GPU)**  
  Критически важна для рендеринга графики.  
  Минимум: NVIDIA RTX 3060 / AMD RX 6600 XT для комфортной игры на средних-высоких настройках в 1080p.

- **Оперативная память (RAM)**  
  Недостаток ОЗУ вызывает лаги и просадки FPS.  
  Минимум: 16 ГБ DDR4 3200 МГц (рекомендуется 32 ГБ для будущего).

- **Накопитель (Storage)**  
  HDD критически замедляет загрузку игр и уровней.  
  Минимум: 1 ТБ NVMe SSD.

- **Блок питания (PSU)**  
  Должен покрывать пиковую мощность системы с запасом 20–30%.  
  Рекомендация: сертификат 80+ Bronze и выше от проверенных брендов (Corsair, Seasonic, be quiet!).

- **Охлаждение**  
  Обеспечивает стабильность под нагрузкой.  
  Минимум: качественный башенный кулер для CPU + 2–3 вентилятора в корпусе.

- **Монитор**  
  Должен соответствовать возможностям видеокарты:  
  — 1080p 144 Гц — для RTX 3060 / RX 6600 XT  
  — 1440p 144 Гц — для RTX 4070 / RX 7800 XT  
  — 4K 60+ Гц — для RTX 4080 / RX 7900 XTX

## 2. Методы и инструменты для проверки

| Инструмент / Ресурс         | Назначение                                                                 |
|----------------------------|----------------------------------------------------------------------------|
| **PCPartPicker**           | Проверка совместимости компонентов, расчёт потребляемой мощности          |
| **UserBenchmark**          | Быстрая оценка производительности относительно среднего рынка             |
| **3DMark / PassMark**      | Стандартизированные бенчмарки для объективного сравнения                  |
| **Steam / Epic Store**     | Официальные системные требования игр                                      |
| **PCGameBenchmark**        | Проверка, запустится ли конкретная игра на данной конфигурации           |
| **GPU-Z / CPU-Z**          | Верификация характеристик «железа» в уже собранном ПК                     |
| **HWMonitor / MSI Afterburner** | Мониторинг температур, загрузки и стабильности под нагрузкой          |

## 4. Пример проверки двух ПК

| Параметр           | ПК, который проходит проверку     | ПК, который не проходит проверку | Критерий прохождения                     |
|--------------------|-----------------------------------|----------------------------------|------------------------------------------|
| Процессор          | AMD Ryzen 7 7800X3D               | Intel Core i3-10100              | Минимум 6 ядер / 12 потоков              |
| Видеокарта         | NVIDIA RTX 4070                   | NVIDIA GT 1030                   | Минимум RTX 3060 / RX 6600 XT            |
| Оперативная память | 32 ГБ DDR5 6000 МГц               | 8 ГБ DDR4 2400 МГц               | Минимум 16 ГБ                            |
| Накопитель         | 2 ТБ NVMe SSD                     | 1 ТБ HDD                         | Обязательно SSD                          |
| Блок питания       | 750 Вт 80+ Gold (Corsair RM750e)  | 400 Вт неизвестного бренда       | Запас мощности 20–30%, сертификат 80+    |
| Итог               | Проходит                       |  Не проходит                   | Все ключевые параметры соответствуют     |

### Пояснение

**ПК, который проходит проверку**  
Конфигурация на базе Ryzen 7 7800X3D и RTX 4070 обеспечит стабильные 90–120 FPS в большинстве современных игр (Cyberpunk 2077, Elden Ring, Hogwarts Legacy) на высоких настройках в 1440p. 32 ГБ ОЗУ и быстрый NVMe SSD исключают бутылочные горлышки при многозадачности и загрузке уровней.

**ПК, который не проходит проверку**  
4-ядерный i3-10100 и видеокарта GT 1030 неспособны обеспечить комфортный геймплей даже на минимальных настройках в большинстве современных игр. 8 ГБ ОЗУ вызовет постоянные лаги, а HDD приведёт к минутным загрузкам уровней. Такая система подходит разве что для казуальных или очень старых игр.
